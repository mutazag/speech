```python
from dotenv import load_dotenv
import os

load_dotenv()
```

```python
from IPython.display import Audio, display
```

```python
region = os.getenv("region") 
key = os.getenv("key")
endpoint = os.getenv("endpoint")
resource = os.getenv("resource")

print(f"Region: {region}")
print(f"Endpoint: {endpoint}")
print(f"Resource: {resource}")
```

```python
import azure.cognitiveservices.speech as speechsdk
```

```python
speech_config = speechsdk.SpeechConfig(subscription=key, region=region)
audio_config = speechsdk.audio.AudioOutputConfig(use_default_speaker=True)
```

```python
speech_config.speech_synthesis_voice_name = "en-US-AvaMultilingualNeural"
```

```python
speech_synthesizer = speechsdk.SpeechSynthesizer(
    speech_config=speech_config, 
    audio_config=audio_config, 
    
)
```

```python
result = speech_synthesizer.speak_text("Hello, how are you today?")
```

```python
print(result.reason)
print(len(result.audio_data))
```

```python
for reason in speechsdk.ResultReason:
      print(reason)
```

```python
if result.reason == speechsdk.ResultReason.SynthesizingAudioCompleted:
    print("Speech synthesized to speaker for text [Hello, how are you today?]")
    display(Audio(data=result.audio_data, rate=16000, autoplay=True))
```
